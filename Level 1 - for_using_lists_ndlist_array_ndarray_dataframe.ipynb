### 1) for loop on Lists

```python
### For using lists
##### Enumerate has the builtin counter for looping through
enumerate(list_name, *kargs, **)
```

```python
print("for loop for lists")
family_heights=[5.6,5.7,5.8,5.4]
for height in family_heights:
    print(height, end = " ")
    #print(height)
```

##### Using Enumerate with for THE same list 

```python
for counter,value in enumerate(family_heights,0): #0 indicates index start, try 2 also
    print("\n index " + str(counter) + ": " + str(value))
print("Here counter is the index, value is the list item at index")
```

##### for loop for list of lists (nested lists)

```python
#nested lists do the slicing just like strings. use square brackets.
#try room[3][1] and house[3,1]
house = [["hall area", 600],
        ["kitchen", 300],
        ["Master Bed",400],
        ["Child Bed",300]]
for room in house:
    #print("The area of " + str(room[0]) + " is " + str(room[1]))
    print(room[0],"size is ", room[1])
print(house[0][1]) # slicing on the list variable
print("use of enumerate on nested lists")
for index,val in enumerate(house):
    print(index,val)
    
```

### 3) for loop on NumPy array

##### 1d numpy array - create array from list

```python
import numpy as np
sales = ['300m','400m','500m']
np_sales = np.array(sales)
for sale in np_sales:
    print(sale)
```

##### 2d array - use of nditer() method to loop through

```python
family_heights1 = [155,170,40,130,150,165]
family_weights = [65,70,28,25,52,64]
#d_arr_hei_wei = np.array(family_heights1,family_weights) 
# this wont work, we have to use np_1d array variables. list variables not supported
np_family_hei = np.array(family_heights1)
np_family_wei = np.array(family_weights)
twoD_arr_hei_wei = np.array([np_family_hei,np_family_wei]) #syntax for 2d array
for arr in np.nditer(twoD_arr_hei_wei):
    print(arr)
#twoD_arr_hei_wei
type(family_heights1)
```
